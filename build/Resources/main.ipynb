```python
# To import pandas
import pandas as pd
```

```python
# Store filepath in a variable
file_path = "Resources/cities.csv"
```

```python
# Read the modified GoodReads csv and store into Pandas DataFrame
cities_df = pd.read_csv(file_path, encoding="utf-8")
cities_df = cities_df.rename(columns={'Unnamed: 0': 'City_ID'})
cities_df
```

```python
cities_df = cities_df.set_index(cities_df['City_ID'])
cities_df = cities_df.drop(columns='City_ID')
cities_df
```

```python
# render dataframe as html
html = cities_df.to_html()
# print(html)
```

```python
# write html to file
text_file = open("index.html", "w")
text_file.write(html)
text_file.close()
```
